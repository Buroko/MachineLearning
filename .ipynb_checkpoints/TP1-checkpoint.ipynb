# Exercice 1 : Model Linéaire (Régression Polynomiale : 1 variable)

Etape 1 : Installation et Importation les librairies

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
import seaborn as sns
```

```python
# sets the color palette for Seaborn plots to use default color codes. Seaborn is a popular data visualization library in Python
#that is built on top of Matplotlib and provides additional customization options.
sns.set(color_codes=True) 

# sets the size of the figure in inches. In this case, the width and height are both
# set to 12 inches, which will create a square-shaped figure.
plt.rcParams["figure.figsize"] = [12,12]
```

Etape 2 : Créer un Dataset

```python
np.random.seed(0) # pour toujours reproduire le meme dataset
x, y = make_regression(n_samples=100, n_features=1, noise=10) #module generant dataset de 100 échantillons, each with 1 feature
plt.scatter(x, y) # afficher les résultats. x en abscisse et y en ordonnée
plt.xlabel('Input feature')
plt.ylabel('Output')
```

```python
print(x.shape)
print(y.shape)
# redimensionner y
y = y.reshape(y.shape[0], 1)
print(y.shape)
```

Etape 3 : Création du modèle et l’entraîner

```python
model = LinearRegression()        #
model.fit(x,y)                    #The fit() method is used to train the model on the data
```

Etape 4 : Evaluation

```python
y_predict = model.predict(x)
plt.scatter(x, y)
plt.plot(x, y_predict, color='g')
plt.show()
print('Coeff R2 =', model.score(x, y))
```

# Exercice 2 : Model non-linéaire (Régression Polynomiale: 1 variable)


Etape 1 : Importer les librairies

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
import seaborn as sns
from sklearn.preprocessing import PolynomialFeatures
import operator

sns.set(color_codes=True)
plt.rcParams["figure.figsize"] = [12,12]
```

Etape 2 : Créer un Dataset

```python
np.random.seed(0) # permet de reproduire l'aléatoire
x, y = make_regression(n_samples=100, n_features=1, noise = 10) # creation d'un dataset (x, y) linéaire
y = y**2
plt.scatter(x, y) # afficher les résultats. x en abscisse et y en ordonnée
```

```python
# On ajoute des variables polynômiales dans notre dataset 
poly_features = PolynomialFeatures(degree=2, include_bias=False) #include_bias : ajout de la ct ou non
x_poly = poly_features.fit_transform(x)  #
```

Etape 3 : Création du modèle et l’entraîner

```python
model = LinearRegression()
model.fit(x_poly, y)
```

Etape 4 : Evaluation

```python
y_poly_pred = model.predict(x_poly)
plt.scatter(x, y, s=10) #cette ligne crée un nuage de points des points de données d'origine, où x est
#la variable prédictive et y est la variable de réponse.

# sort the values of x before line plot
sort_axis = operator.itemgetter(0)
sorted_zip = sorted(zip(x,y_poly_pred), key=sort_axis)
x_p, y_poly_pred_P = zip(*sorted_zip)
plt.plot(x_p, y_poly_pred_P, color='g')
plt.show()
print('Coeff R2 =', model.score(x_poly, y))
```

# Exercice 4 : Régression linéaire sur un vrai Dataset

# goal : prédiction des prix de maison

Etape1 : Importer les librairies nécessaire

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
import seaborn as sns

from sklearn.preprocessing import PolynomialFeatures
import operator
 
from sklearn.model_selection import train_test_split
```

Etape 2 : charger Dataset et Prétraitement Boston Housing Dataset

```python
path = 'hou_all.csv'
data = pd.read_csv(path, index_col=0, encoding = "ISO-8859-1")
```

```python
data.head(2)
```

```python
# CRIM  : taux de criminalité par habitant et par ville
# ZN    : Proportion des terrains résidentiels zonés pour les terrains de plus de 25 000 sq. ft
# INDUS : Proportion d'acres commerciaux non commerciaux par ville
# CHAS  : variable fictive Charles River (= 1 si le secteur délimite la rivière ; 0 sinon)
# NOX   : Concentration d'oxyde nitrique (parties par 10 millions)
# RM    : Nombre moyen de pièces par logement
# AGE   : Proportion de logements occupés par leur propriétaire construits avant 1940
# DIS   : Distances pondérées par rapport à cinq centres d'emploi de Boston
# RAD   : Indice d'accessibilité aux autoroutes radiales
# TAXE  : Taux de la taxe foncière sur la valeur totale par tranche de 10 000 $
# PTRATIO : Ratio élèves-enseignant par commune
# B     : 1000(Bk — 0,63)², où Bk est la proportion de [personnes d'ascendance afro-américaine] par ville
# LSTAT : Pourcentage de statut inférieur de la population
# MEDV  : Valeur médiane des maisons occupées par leur propriétaire en milliers de dollars
# Important :Les prix de la maison indiqués par la variable MEDV sont notre variable cible 
# (target) et les autres sont les variables caractéristiques (features) sur la base desquelles nous 
# prédirons la valeur d'une maison.
```

```python
data.zn.replace(0,np.nan,inplace = True)
data.chas.replace(0,np.nan,inplace=True)
data.info()
```

```python

data[data == 0] = np.nan
data.head()
```

```python
data.isnull().sum()/len(data) * 100
```

```python
data = data.drop(['crim','indus'],axis=1)
data.info()
```

```python
data.head()
```

L'analyse exploratoire des données

```python
sns.set(rc={'figure.figsize':(11,8)})
sns.distplot(data['zn'], bins=30)
plt.show()
```

```python
#Nous utiliserons la fonction heatmap de la bibliothèque seaborn pour tracer la matrice de corrélation
correlation_matrix = data.corr().round(2)
sns.heatmap(data=correlation_matrix, annot=True) # annot true:afficher numeros dans les case , false hide them
```

```python
#Que remarquez-vous ? => c'est une matrice medianne et les features n'ont aucune relation entre eux
```

Division des données en ensembles train et de test

```python
#Pour diviser les données, nous utilisons la fonction train_test_split
X = data.iloc[:,:-1]
y = data.iloc[:,-1]
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=0)
```

Etape 3 : Création du modèle et l’entraîner


```python
 
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

Etape 4 : Model evalution


```python
 
from sklearn.metrics import r2_score
y_pred = model.predict(X_test)
from sklearn import metrics 
print("MAE", metrics.mean_absolute_error(y_test, y_pred))
print("MSE", metrics.mean_squared_error(y_test, y_pred))
print("RMSE", np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
print("Score:", model.score(X_test, y_test))
print("r2_score:", r2_score(y_test,y_pred))
```
